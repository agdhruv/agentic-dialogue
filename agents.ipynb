```python
from autogen import ConversableAgent, GroupChat, GroupChatManager
from constants import config_list

# Define the admin agent
admin = ConversableAgent(
    name="Admin",
    system_message="You are a neutral discussion administrator. Your role is to pose the initial question and encourage debate among the agents. Once a consensus is reached, say TERMINATE.",
    llm_config={"config_list": config_list},
    is_termination_msg=lambda msg: msg.get("content") is not None and "TERMINATE" in msg.get("content")
)

# Define a single discussion agent
agent1 = ConversableAgent(
    name="Agent1",
    system_message="You are an agent in the discussion. Engage in debate. Output no more than a line. Be willing to change your mind.",
    llm_config={"config_list": config_list},
    description="An agent in the discussion.",
    is_termination_msg=lambda msg: msg.get("content") is not None and "TERMINATE" in msg.get("content")
)

# Define a single discussion agent
agent2 = ConversableAgent(
    name="Agent2",
    system_message="You are an agent in the discussion. Engage in debate. Output no more than a line. Be willing to change your mind.",
    llm_config={"config_list": config_list},
    description="An agent in the discussion.",
    is_termination_msg=lambda msg: msg.get("content") is not None and "TERMINATE" in msg.get("content")
)

# Set up the group chat with just one agent for simplicity
group_chat = GroupChat(
    agents=[admin, agent1, agent2],
    messages=[],
    max_round=10,
    send_introductions=True,
)

# Set up the manager
manager = GroupChatManager(
    groupchat=group_chat,
    llm_config={"config_list": config_list},
    is_termination_msg=lambda msg: msg.get("content") is not None and "TERMINATE" in msg.get("content")
)
```
